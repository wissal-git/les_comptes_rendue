###  EXERCICE 1 :


 Soit une fonction f (inconnue explicitement) connue seulement en certains points x0,x1…xn  ou évaluable par un calcul coûteux.
- Principe :  représenter f par une fonction simple, facile à évaluer

- Problème :   il existe une infinité de solutions !
 
 **la premiére code permet de calculer le polynome substituant de la fonction f :**

```python

from numpy import prod

#import math
def PR(x,y,c):
    S=[]
    n=len(x)
    l=[]
    for i in range(n) :
        l=[c-i  for i in x] 
            
        for j in range(n):
            if (i==j):
                l[j]=1
            else :
                l[j]=l[j]/(x[i] - x[j])
        S.append(y[i]*prod(l))
    u=sum(S)
    return u
    
X=[1 ,2, 3 ,4]
Y=[-1, 0, 2 ,1]
for i in X :
    print (PR(X,Y,i))
#print(l)    
```

     **p(xi) = yi pour i dans  [0,n]**

2) Les points de Tchebycheff sur un intervalle $[a;b]$ sont définie par :
  $$x_k=\dfrac{a+b}{2}+\dfrac{a-b}{2}\cos\left(\frac{k-1}{n-1}\pi\right)\hspace{0.5cm} (k=1,..,n: x_1=a:x_n=b)$$

  Vérifier que sur $[-5; 5]$ ; l'interpolation polynômiale de $f(x) = \dfrac{1}{1 + x^2}$ ; en $N$ points de Tchebychev
est meilleure qu'une  interpolation utilisant $N$ points uniforment espacés (on prendra $N=10, 11, 16, 17$).

<h3>
<span style="color:indigo"> 
Interpolation de Tchebychev
    </span>
    </h3>
     
   

```python
def tcheb(a,b,n):
    v=[]
    for k in range(n):
        v.append((a+b)/2+(a-b)/2*math.cos((k-1)/(n-1)*math.pi))
    return v
Xt=tcheb(-5,5,11)
print(Xt)
```

```python
import matplotlib.pyplot as plt
from numpy import linspace 
import numpy as np



N=10
#X=linspace(-5,5,N)
X=np.array(tcheb(-5,5,N))
f=lambda x: 1/(1+x**2)
Y=f(X)
p=np.polyfit(X,Y,N-1)
t=linspace(-5,5,1000)

plt.plot(t,f(t),'r',X,Y,'ob', t, np.polyval(p,t),'g')
plt.title("Interpolation Equidistante avec N=10")

plt.show()
##########################################
N=11
#X=linspace(-5,5,N)
X=np.array(tcheb(-5,5,N))
f=lambda x: 1/(1+x**2)
Y=f(X)
p=np.polyfit(X,Y,N-1)
t=linspace(-5,5,1000)

plt.plot(t,f(t),'r',X,Y,'ob', t, np.polyval(p,t),'g')
plt.title("Interpolation Equidistante avec N=11")
plt.show()
##########################################
N=16
#X=linspace(-5,5,N)
X=np.array(tcheb(-5,5,N))
f=lambda x: 1/(1+x**2)
Y=f(X)
p=np.polyfit(X,Y,N-1)
t=linspace(-5,5,1000)

plt.plot(t,f(t),'r',X,Y,'ob', t, np.polyval(p,t),'g')
plt.title("Interpolation Equidistante avec N=16")

plt.show()

##########################################
N=16
#X=linspace(-5,5,N)
X=np.array(tcheb(-5,5,N))
f=lambda x: 1/(1+x**2)
Y=f(X)
p=np.polyfit(X,Y,N-1)
t=linspace(-5,5,1000)

plt.plot(t,f(t),'r',X,Y,'ob', t, np.polyval(p,t),'g')
plt.title("Interpolation Equidistante avec N=17")

plt.show()
```
